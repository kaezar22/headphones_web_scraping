```python
import requests 
from bs4 import BeautifulSoup
from csv import writer
import pandas as pd
url="https://listado.mercadolibre.com.co/audifonos_Desde_51_NoIndex_True"
page = requests.get(url)
print(page)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
lists = soup.find_all('div',class_ = "ui-search-result__wrapper shops__result-wrapper")
stars = soup.find('span',class_ = 'ui-search-reviews__ratings')
```

```python
#write a csv
#'star2','star3','star4','star5',star2,star3,star4,star5,
with open('prueba.csv', 'w', encoding = 'utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title', 'Price','Reviews','star1','star2','star3','star4','star5']
    thewriter.writerow(header)
    for list in lists:
        title = list.find('h2',class_ = "shops__item-title").text
        price = list.find('span',class_ = "price-tag-fraction").text
        reviews = list.find('span',class_= "ui-search-reviews__amount")
        for star in stars:
            star.use
            
        
        info = [title,price,reviews,star1,star2,star3,star4,star5]
        thewriter.writerow(info)
```

```python
#iterate in multiple pages of the same search and, produce a csv
headphones_final = []
for i in range(51,1051,50):
    url = f"https://listado.mercadolibre.com.co/audifonos_Desde_{i}_NoIndex_True"
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    lists = soup.find_all('div',class_ = "ui-search-result__wrapper shops__result-wrapper")
    stars = soup.find('span',class_ = 'ui-search-reviews__ratings')

    for list in lists:
        title = list.find('h2',class_ = "shops__item-title").text
        price = list.find('span',class_ = "price-tag-fraction").text
        reviews = list.find('span',class_= "ui-search-reviews__amount")
        for star in stars:
            star.use
        headphones_final.append([title,price,reviews,star1,star2,star3,star4,star5])
           
```

```python
df = pd.DataFrame(headphones_final,columns = ['Title','Price','Reviews','star1','star2','star3','star4','star5'])
df.head()
```

```python
df.to_csv('headphones_final.csv')
```
